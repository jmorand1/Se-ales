# $\Rightarrow$ Encuentre el valor de $\omega_o\in \mathbb{R}$ para que el conjunto $\{e^{jnω_ot}\}_{n\in\mathbb{Z}}$ sea ortogonal en el intervalo $t\in[−T /2, T /2]$,con $T\in R^+$.



Para demostrar la ortogonalidad se toman dos frecuencias diferentes y se las evalúa en el intervalo requerido de la siguiente manera:

$$\int_{-T/2}^{T/2}e^{jn\omega_ot}e^{-jm\omega_ot}dt$$

Con $n\not= m$, de lo contrario, el resultado de la integral es $T$.

$$\int_{-T/2}^{T/2}e^{jn\omega_ot}e^{-jm\omega_ot}dt = \int_{-T/2}^{T/2}e^{(n-m)j\omega_ot}dt$$

$$\int_{-T/2}^{T/2}e^{jn\omega_ot}e^{-jm\omega_ot}dt = \frac{1}{(n-m)jw_o}e^{(n-m)jw_ot}\Big|_{-T/2}^{T/2}$$

Ahora se utiliza la sustitución $r = n-m$

$$\int_{-T/2}^{T/2}e^{jn\omega_ot}e^{-jm\omega_ot}dt = \frac{1}{jrw_o}[e^{jrw_o\frac{T}{2}}-e^{-jrw_o\frac{T}{2}}]$$

En este punto ya se puede identificar el valor de $\omega_o$ dado que esta integral debe ser cero, así que si hacemos la sustitución:

$$w_o=\frac{2\pi}{\text{Intervalo de análisis}}=\frac{2\pi}{T}$$

Para después transformar los euler complejos en un seno así:

$$\frac{1}{jrw_o}[e^{jrw_o\frac{T}{2}}-e^{-jrw_o\frac{T}{2}}]=\frac{1}{jrw_o}[e^{jr\frac{2\pi}{T}\frac{T}{2}}-e^{-jr\frac{2\pi}{T}\frac{T}{2}}]=2\frac{sen(r\pi)}{rw_o} = 0$$

De tal modo que si $w_o = \frac{2\pi}{T}$ el conjunto $e^{jn\omega_ot}$ es ortogonal.

 # $\Rightarrow$ Demuestre la ortogonalidad del conjunto $\{\cos(n\omega_ot),\sin (m\omega_ot)\} n,m\in\mathbb{Z}.$


Para comprobar ortagonalidad en el conjunto se necesita hacer 6 integrales que son:

1.   $$\int_{-T/2}^{T/2}cos(n\omega_o t)cos^*(m\omega_o t)dt = \int_{-T/2}^{T/2}cos(n\omega_o t)cos(m\omega_o t)dt$$
2.   $$\int_{-T/2}^{T/2}sen(n\omega_o t)sen^*(m\omega_o t)dt = \int_{-T/2}^{T/2}sen(n\omega_o t)sen(m\omega_o t)dt$$
3.   $$\int_{-T/2}^{T/2}cos(n\omega_o t)sen^*(m\omega_o t)dt = \int_{-T/2}^{T/2}cos(n\omega_o t)sen(m\omega_o t)dt$$

Cada una para cuando $n=m$ y $n\not=m$

1.  $$\int_{-T/2}^{T/2}cos(n\omega_o t)cos^*(m\omega_o t)dt$$

*  $n = m$

   $$\int_{-T/2}^{T/2}cos(n\omega_o t)cos(n\omega_o t)dt = \int_{-T/2}^{T/2}cos^2(n\omega_o t)dt$$
   $$=[\frac{1}{2}t + \frac{1}{2(2n\omega_o t)}[sen(2n\omega_o t)]\Big|_{-T/2}^{T/2}=\frac{T}{2}= \varepsilon$$

*  $n \not= m$

   $$\int_{-T/2}^{T/2}cos(n\omega_o t)cos(m\omega_o t)dt$$
   Se utiliza la identidad trigonometrica:
   $$cos(\alpha)cos(\beta)= \frac{1}{2}[cos(\alpha + \beta)+cos(\alpha-\beta)] $$

   $$\int_{-T/2}^{T/2}cos(n\omega_o t)cos(m\omega_o t)dt = \frac{1}{2}\bigg[\int_{-T/2}^{T/2}cos((n+m)\omega_o t)dt+\int_{-T/2}^{T/2}cos((n-m)\omega_o t)dt\bigg]$$

   $$\text{Se sustituye: } r_1 = n+m; r_2 = n-m$$

   $$=\frac{1}{2}\bigg[\int_{-T/2}^{T/2}cos(r_1\omega_o t)dt+\int_{-T/2}^{T/2}cos(r_2\omega_o t)dt\bigg]$$

   $$= \frac{1}{2}(0+0) = 0$$

2.  $$\int_{-T/2}^{T/2}sen(n\omega_o t)sen(m\omega_o t)dt$$



*  $n = m$

   $$\int_{-T/2}^{T/2}sen(n\omega_o t)sen(n\omega_o t)dt = \int_{-T/2}^{T/2}sen^2(n\omega_o t)dt$$

   $$=[\frac{1}{2}t-\frac{1}{4(n\omega_o)}sen(2n\omega_o t)]\bigg|_{-T/2}^{T/2}=\frac{T}{2}=\varepsilon$$

*  $n \not= m$
   
   $$\int_{-T/2}^{T/2}sen(n\omega_o t)sen(m\omega_o t)dt$$

   Se utiliza la identidad trigonométrica:

   $$sen(\alpha)sen(\beta)= \frac{1}{2}[-cos(\alpha + \beta)+cos(\alpha-\beta)]$$

   $$\int_{-T/2}^{T/2}cos(n\omega_o t)cos(m\omega_o t)dt = \frac{1}{2}\bigg[-\int_{-T/2}^{T/2}cos((n+m)\omega_o t)dt+\int_{-T/2}^{T/2}cos((n-m)\omega_o t)dt\bigg]$$

   $$\text{Se sustituye: } r_1 = n+m; r_2 = n-m$$

   $$=\frac{1}{2}\bigg[-\int_{-T/2}^{T/2}cos(r_1\omega_o t)dt+\int_{-T/2}^{T/2}cos(r_2\omega_o t)dt\bigg]$$

   $$= \frac{1}{2}(-0+0) = 0$$ 

3. $$\int_{-T/2}^{T/2}cos(n\omega_o t)sen(m\omega_o t)dt$$

*  $n = m$

   $$\int_{-T/2}^{T/2}cos(n\omega_o t)sen(n\omega_o t)dt = [-\frac{1}{4n\omega_o}cos(2n\omega_o t)]\bigg|_{-T/2}^{T/2}=0\not=\varepsilon$$

*  $n \not= m$

   $$\int_{-T/2}^{T/2}cos(n\omega_o t)sen(m\omega_o t)dt$$

   Se utiliza la identidad trigonométrica:

   $$sen(\alpha)cos(\beta)= \frac{1}{2}[sen(\alpha + \beta)+sen(\alpha-\beta)]
   $$

   $$\int_{-T/2}^{T/2}sen(n\omega_o t)cos(m\omega_o t)dt= \frac{1}{2}\bigg[\int_{-T/2}^{T/2}sen((n+m)\omega_o t)dt+\int_{-T/2}^{T/2}sen((n-m)\omega_o t)dt\bigg]
   $$

   $$\int_{-T/2}^{T/2}sen(n\omega_o t)cos(m\omega_o t)dt = \frac{1}{2}(0+0) = 0$$

 # $\Rightarrow$ Encuentre el espectro $c_n \in C$ de la señal $x(t) = u(t)$ a través de la representación generalizada $x(t) = \sum_{n=−\infty}^{\infty} c_n\phi_n(t)$,para $\phi_n(t) = e^{jnω_ot}$. Realice una simulacion en Python para $t \in [0, 2]$ y determine el error relativo para $n \in \{0, \pm1, \pm2\}$.

Nótese que al utilizar $\phi_n(t) = e^{jn\omega_o t}$ se está utilizando las bases de Fourier por lo tanto tenemos que para encontrar $c_n$ se utiliza la siguiente ecuación:

$$c_n = \frac{1}{T}\int_{-T/2}^{T/2}x(t)e^{-jn\omega_o t}dt$$

De este modo:

$$c_n = \frac{1}{T}\int_{-T/2}^{T/2}u(t)e^{-jn\omega_o t}dt$$

$$c_n = \frac{1}{T}\int_{0}^{T/2}e^{-jn\omega_o t}dt=\frac{1}{T}\big[ \frac{e^{-\frac{T}{2}jn\omega_o}-1}{-jn\omega_o}\big]$$

$$c_n = -\frac{1}{T}\big[\frac{e^{-jn\pi}-1 }{jn\omega_o}\big] = j[\frac{e^{-jn\pi}-1}{2n\pi}]$$

Con ayuda de la ecuación: $e^{-jn\pi}=(-1)^n$

$$c_n = j[\frac{(-1)^n-1}{2n\pi}]$$



```
# Para graficar
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

T = 4 #3 segundos de grafica
Fs = 500 # 500 muestras por segundo
Nm = T*Fs #numero de muestras requeridas
tv = np.linspace(-T/2,T/2,Nm)

xo = np.concatenate((np.zeros(int(Nm/2)), np.ones(int(Nm/2))), axis=0)

plt.plot(tv,xo,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("x(t)",fontsize = 14)
```

```
N = 200 #numero armonicos
wo = 2*np.pi/T # frecuencia fundamentas
# %% definir bases
phin = np.zeros((Nm,2*N+1),dtype=np.complex_)
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)
```

Se calcula $c_o$

$$c_o = \frac{1}{T}\int_{-T/2}^{T/2}x(t)dt=\frac{1}{T}\int_{-T/2}^{T/2}u(t)dt = \frac{1}{2}$$

```
cn = np.zeros(2*N+1,dtype=np.complex_)
nv = np.linspace(-N, N, 2*N+1) #vector num armónicos
Cn = ((1j)/(2*nv*np.pi))*((np.exp((-1j)*nv*np.pi))-1) 
Cn[N] = 1/2 
#graficar espectro
#plt.figure(figsize = (10, 8))
plt.stem(nv, abs(Cn), 'r')
plt.title(r'|Cn|')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$|C_n|$',fontsize = 14)
plt.axis('tight')
plt.grid()
```

```
xe = phin @ Cn  #producto matricial para calcular componentes espectrales phin.dot(Cn)
plt.plot(tv,xe,label = '$\hat{x}(t)$')
plt.plot(tv,xo,label = '$x(t)$')
plt.plot(tv,xo-xe,label = '$e(t)$')
plt.legend()
```

```
Np = 3 #numero de periodos a graficar
tv2 = np.linspace(-Np*T/2,Np*T/2,Np*T*Fs)
phin2 = np.zeros((len(tv2),2*N+1),dtype=np.complex_)
for n in range(-N,N+1,1):
    phin2[:,n+N] = np.exp(1j*n*wo*tv2)
    
xe2 = phin2 @ Cn  #producto matricial para calcular componentes espectrales
plt.plot(tv2,xe2)
```

La ecuación que define el error relativo es la siguiente:

$$er = 1-\frac{|c_o|^2+2\sum_{n = 1}^{N}|c_n|^2}{P_x}$$

con $P_x = \frac{1}{T}\int_{-T/2}^{T/2}|x(t)|^2dt=\frac{1}{T}\int_{-T/2}^{T/2}u(t)dt = \frac{1}{2}$

$c_{1} = \frac{1}{j\pi};c_{-1} = \frac{-1}{j\pi} ; c_{\pm2} = 0$

$$|c_{\pm1}|^2=\frac{1}{\pi^2} $$

$$er= 1-\frac{|\frac{1}{2}|^2+2(\frac{1}{\pi^2}+0)}{\frac{1}{2}}=0.09471$$

$$\text{En porcentaje: } 9.471\%$$

# $\Rightarrow$Sea $x''(t)$ la segunda derivada de la señal $x(t)$, donde $t \in [ti  , tf ]$. Demuestre que los coeficientes de la serie exponencial de Fourier se pueden calcular segun: $c_n = \frac{1}{ (ti − tf)n^2w_o^2}\int_{ti}^{tf} x''(t)e^{−jnw_ot}dt; n \in \mathbb{Z}$. 

#Como se pueden calcular los coeficientes  $a_n$ y $b_n$ desde $x''(t)$ en la serie trigonometrica de Fourier? 

Podemos identificar que $\int_{ti}^{tf} x''(t)e^{−jnw_ot}dt$ corresponde a $\mathbb{F}\{\frac{d^2}{dt^2}x(t)\}=(jn\omega_o)^2\int_{ti}^{tf}x(t)e^{-jn\omega_ot}dt$ por lo tanto, remplazando en la ecuación inicial:

$$c_n = \frac{(jn\omega_o)^2}{ (ti − tf)n^2w_o^2}\int_{ti}^{tf}x(t)e^{−jnw_ot}dt$$

$$c_n = -\frac{n^2\omega_o^2}{(ti-tf)n^2\omega_o^2}\int_{ti}^{tf}x(t)e^{−jnw_ot}dt$$

$$c_n = \frac{1}{(tf-ti)}\int_{ti}^{tf}x(t)e^{−jnw_ot}dt$$

Ahora, haciendo $tf-ti = T$

$$c_n = \frac{1}{T}\int_{ti}^{tf}x(t)e^{−jnw_ot}dt$$

Que es la fórmula de $c_n$ para Fourier.




Para determinar $a_n$ y $b_n$ en términos de $x''(t)$ lo que se hace es descomponer el euler complejo de la fórmula de $c_n$ que ya sabemos es correcta:

$$c_n = \frac{1}{ (ti − tf)n^2w_o^2}\int_{ti}^{tf} x''(t)e^{−jnw_ot}dt$$ 

$$c_n = \frac{1}{ (ti − tf)n^2w_o^2}\int_{ti}^{tf} x''(t)[cos(n\omega_ot)-jsen(n\omega_ot)]dt$$

Ahora, se sabe que:

$a_n = 2*Re\{c_n\}$

$b_n = -2*Im\{c_n\}$

Por lo tanto $a_n$ y $b_n$ en términos de $x''(t)$:

$$a_n = \frac{2}{ (ti − tf)n^2w_o^2}\int_{ti}^{tf} x''(t)cos(n\omega_ot)dt$$

$$b_n = \frac{2}{ (ti − tf)n^2w_o^2}\int_{ti}^{tf} x''(t)sen(n\omega_ot)dt$$

# $\Rightarrow$¿Como se puede simplificar el cálculo de la serie trigonométrica de Fourier para señales con simetría de media onda y de cuarto de onda?. 

#En dichos casos, ¿como se pueden calcular los coeficientes $a_n$ y $b_n$ con $n \in \mathbb{N}$?

$\Rightarrow$Simetría de media onda:

Que una señal tenga simetría de media onda significa que se ha corrido medio periodo y se ha invertido. La función que representa esto es:
$$f(t) = -f(t-\frac{T}{2})$$

Las ecuaciones simplificadas de la serie trigonométrica de Fourier son las siguientes:

*   Para $n$ par:

    $a_n = b_n = 0$

*   Para $n$ impar:

    $a_n = \frac{4}{T}\int_0^{T/2}f(t)cos(n\omega_o t)dt$

    $b_n = \frac{4}{T}\int_{t_o}^{T/2}f(t)sen(n\omega_o t)dt$

$\Rightarrow$Simetría de cuarto de onda:

Que una señal tenga simetría de cuarto de onda significa que tiene simetría de media onde y además es función par o impar.

Las ecuaciones simplificadas de la serie trigonométrica de Fourier son las siguientes:

*   Simetría de media onda y función par con $n$ impar:

    $a_n = \frac{8}{T}\int_0^{T/4}f(t)cos(n\omega_o t)dt$

    El resto de valores son cero

*   Simetría de media onda y función impar con $n$ impar:

    $a_n = \frac{8}{T}\int_{t_o}^{T/4}f(t)sen(n\omega_o t)dt$

    El resto de valores son cero

$\Rightarrow$ Relaciones entre serie y transformada de Fourier

*   La serie de Fourier es una expansión de la señal periódica como una combinación lineal de senos y cosenos, mientras que la transformada de Fourier es el proceso o la función utilizada para convertir las señales del dominio del tiempo en el dominio de la frecuencia.

*   La serie de Fourier se define para señales periódicas y la transformada de Fourier se puede aplicar a señales aperiódicas (que ocurren sin periodicidad).

*   Las series de Fourier descomponen una función periódica en una suma de senos y cosenos con diferentes frecuencias y amplitudes, mientras que la transformada de Fourier es una operación matemática que rompe una señal en sus frecuencias constituyentes.

*   La serie de Fourier va de tiempo contínuo a frecuencia discreta, mientras que la transformada de Fourier va de tiempo contínuo a una frecuencia contínua, esto es:

  $t \in \{ -T/2,T/2 \}\rightarrow n\in \mathbb{Z}$

  $t \in \{ -\infty,\infty \}\rightarrow \omega\in \{-\infty,\infty\}$








######$\Rightarrow$ Fast Fourier Transform (FFT)
La transformada rápida de Fourier, conocida por la abreviatura FFT (del inglés Fast Fourier Transform) es un algoritmo eficiente que permite calcular la transformada de Fourier discreta (DFT) y su inversa. La FFT es de gran importancia en una amplia variedad de aplicaciones, desde el tratamiento digital de señales y filtrado digital en general a la resolución de ecuaciones en derivadas parciales o los algoritmos de multiplicación rápida de grandes enteros. Cuando se habla del tratamiento digital de señales, el algoritmo FFT impone algunas limitaciones en la señal y en el espectro resultante ya que la señal muestreada y que se va a transformar debe consistir de un número de muestras igual a una potencia de dos. La mayoría de los analizadores de FFT permiten la transformación de 512, 1024, 2048 o 4096 muestras. El rango de frecuencias cubierto por el análisis FFT depende de la cantidad de muestras recogidas y de la proporción de muestreo.

##a) $t$

```
# Para graficar
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

T = 4 #3 segundos de grafica
Fs = 500 # 500 muestras por segundo
Nm = T*Fs #numero de muestras requeridas
tv = np.linspace(-T/2,T/2,Nm)

xo = tv

plt.plot(tv,xo,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("x(t)",fontsize = 14)
```

```
#cálculo de la FFT de la señal signo invertido
vf = np.fft.fftfreq(len(xo),1/Fs) #calculo vector de frecuencias
Xw = np.fft.fft(xo) #cálculo FFT

plt.stem(vf,abs(Xw),'r')
plt.title(r'|X[n]|')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|$',fontsize = 14)
plt.axis('tight')
plt.grid()
plt.show()
plt.stem(vf,abs(Xw)/len(vf),'r')
plt.title(r'|X[n]|/K=c_n')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|/K$',fontsize = 14)
```

```
#dado que el espectro es simétrico, y aprovechando señales netamente reales
#se sugiere utilizar la expresión simplificada rfft

vf = np.fft.rfftfreq(len(xo),1/Fs) #calculo vector de frecuencias
Xw = np.fft.rfft(xo) #cálculo FFT
plt.stem(vf,abs(Xw),'r')
plt.title(r'|X[n]|')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|$',fontsize = 14)
plt.axis('tight')
plt.grid()
plt.show()
plt.stem(vf,abs(Xw)/(2*len(vf)),'r')
plt.title(r'|X[n]|/K=c_n')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|/K$',fontsize = 14)
```

```
#la reconstrucción de x[k] se puede obtener mediante la transformada inversa
xe = np.fft.irfft(Xw)
plt.plot(tv,xe)
plt.title(r'Reconstrucción completa')
plt.xlabel(r't',fontsize = 14)
plt.ylabel(r'$x[n]$',fontsize = 14)
plt.show()
#se pueden escoger o ponderar los coeficientes para filtrar (alterar) la reconstrucción
ind = range(10,750)
Xwf = Xw.copy()
Xwf[ind] = 0
xe2 = np.fft.irfft(Xwf)
plt.plot(tv,xe2)
plt.show()
```

# $\Rightarrow$Encuentre la funcion de densidad espectral (transformada de Fourier) para las siguientes señales: 
#a) $e^{−a|t|}, a\in\mathbb{R}^+$

#b) $cos (\omega_ct), \omega_c\in\mathbb{R}$ 

#c) $sin (\omega_st), \omega_s\in\mathbb{R}$

#d) $f(t) cos (\omega_ct), \omega_c\in\mathbb{R}, f(t)\in \mathbb{R,C}$ 

#e) $e^{−a|t|^2}, a\in\mathbb{R^+}$

a) $x(t)=e^{-a|t|}$

$$x(\omega)=\int_{-\infty}^{\infty}e^{-a|t|}e^{-j\omega t}dt=\int_{-\infty}^{0}e^{t(-a-j\omega)}dt+\int_{0}^{\infty}e^{-t(a+j\omega)}dt$$

$$x(\omega) = \frac{1}{j\omega+a}-\frac{1}{j\omega-a} = \frac{2a}{a^2+\omega^2}$$



b) $cos (\omega_ct)$

$$x(\omega)=\int_{-\infty}^{\infty}cos (\omega_ct)e^{-j\omega t}dt$$

Ahora haciendo $cos(\omega_ct) = \frac{e^{j\omega_ct}+e^{-j\omega_ct}}{2}$

$$x(\omega)=\int_{-\infty}^{\infty}(\frac{e^{j\omega_ct}+e^{-j\omega_ct}}{2})e^{-j\omega t}dt$$

Ahora separando los términos en 2 integrales:

$$x(\omega)=\frac{1}{2}(\int_{-\infty}^{\infty}e^{j\omega_ct}e^{-j\omega t}dt+\int_{-\infty}^{\infty}e^{-j\omega_ct}e^{-j\omega t}dt)$$

Por propiedades de la transformada de Fourier: $\mathbb{F}\{x(t)e^{\mp j\omega_o t}\}=2\pi\delta(-(\omega \pm \omega_o))$

Por lo tanto el resultado sería:

$$x(\omega)=\frac{2\pi\delta(-(\omega-\omega_c))+2\pi\delta(-(\omega+\omega_c))}{2}=\pi[\delta(-(\omega-\omega_c))+\delta(-(\omega+\omega_c))]$$




c) $sin(\omega_st)$

$$x(\omega)=\int_{-\infty}^{\infty}sen(\omega_st)e^{-j\omega t}dt$$

Ahora haciendo $sen(\omega_st) = \frac{e^{j\omega_st}-e^{-j\omega_st}}{2j}$

$$x(\omega)=\int_{-\infty}^{\infty}(\frac{e^{j\omega_st}-e^{-j\omega_st}}{2j})e^{-j\omega t}dt$$

Ahora separando los términos en 2 integrales:

$$x(\omega)=\frac{1}{2j}(\int_{-\infty}^{\infty}e^{j\omega_st}e^{-j\omega t}dt-\int_{-\infty}^{\infty}e^{-j\omega_st}e^{-j\omega t}dt)$$

Por propiedades de la transformada de Fourier: $\mathbb{F}\{x(t)e^{\mp j\omega_o t}\}=2\pi\delta(-(\omega \pm \omega_o))$

Por lo tanto el resultado sería:

$$x(\omega)=\frac{2\pi\delta(-(\omega-\omega_s))+2\pi\delta(-(\omega+\omega_s))}{2j}=\frac{\pi}{j}[\delta(-(\omega-\omega_s))+\delta(-(\omega+\omega_s))]$$


d) $f(t) cos (\omega_ct)$

$$x(\omega)=\int_{-\infty}^{\infty}f(t)cos(\omega_ct)e^{-j\omega t}dt$$

Ahora haciendo $cos(\omega_ct) = \frac{e^{j\omega_ct}+e^{-j\omega_ct}}{2}$

$$x(\omega)=\int_{-\infty}^{\infty}f(t)(\frac{e^{j\omega_ct}+e^{-j\omega_ct}}{2})e^{-j\omega t}dt$$

Ahora separando los términos en 2 integrales:

$$x(\omega)=\frac{1}{2}(\int_{-\infty}^{\infty}f(t)e^{j\omega_ct}e^{-j\omega t}dt+\int_{-\infty}^{\infty}f(t)e^{-j\omega_ct}e^{-j\omega t}dt)$$

Por propiedades de la transformada de Fourier: $\mathbb{F}\{x(t)e^{\mp j\omega_o t}\}=X(\omega\pm\omega_o)$

Por lo tanto el resultado sería:

$$x(\omega)=\frac{1}{2}(F(\omega-\omega_c)+F(\omega+\omega_c))$$


e) $e^{−a|t|^2}$

$$x(\omega)=\int_{-\infty}^{\infty}e^{-a|t|^2}e^{-j\omega t}dt=\int_{-\infty}^{\infty}e^{-at^2}e^{-j\omega t}dt=\int_{-\infty}^{\infty}e^{t(-at-j\omega)}dt$$

Resolviendo la integral se obtiene como resultado:

$$x(\omega)=\frac{\sqrt{\pi}e^{\frac{-\omega^2}{4a}}}{\sqrt{a}}$$

##THD
$\rightarrow$ La distorsión armónica es un parámetro técnico utilizado para definir la señal de audio que sale de un sistema.

La distorsión armónica se produce cuando la señal de salida de un sistema no equivale a la señal que entró en él. Esta falta de linealidad afecta a la forma de la onda, porque el equipo ha introducido armónicos que no estaban en la señal de entrada. Puesto que son armónicos, es decir múltiplos de la señal de entrada, esta distorsión no es tan disonante y es más difícil de detectar.

En todo sistema de audio siempre se produce una pequeña distorsión de la señal, dado que todos los equipos actuales introducen alguna no linealidad.

$\rightarrow$ En sistemas eléctricos de corriente alterna ―igual que en acústica― los armónicos son frecuencias múltiplos de la frecuencia fundamental de trabajo del sistema y cuya amplitud va decreciendo conforme aumenta el múltiplo. En el caso de sistemas alimentados por la red de 50 Hz, pueden aparecer armónicos de 100, 150, 200 Hz o

Cuando se habla de los armónicos en las instalaciones de energía, son los armónicos de corriente los más preocupantes, puesto que son corrientes que generan efectos negativos. Es habitual trabajar únicamente con valores correspondientes a la distorsión armónica total (THD). 

$\rightarrow$ Definimos THD de la siguiente manera:

$$THD = \frac{\sum\text{Potencia de los armónicos}}{\text{Potencia de la frecuancia fundamental}}=\frac{P_1+P_2+...+P_N}{P_0}$$

El Factor de Potencia (PF) puede calcularse en relación con el THD de la siguiente forma:

$$PF = \frac{cos(\phi_i)}{\sqrt{1+THD_i^2}}$$
